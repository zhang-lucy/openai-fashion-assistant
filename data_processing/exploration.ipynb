# Figure out primary key
 Let's figure out what combination of text fields is unique, so we can concatenate + hash to create our primary keys

```python
unique_titles = df["title"].unique()
```

```python
from collections import defaultdict
title_dict = defaultdict(int)
for title in list(df["title"]):
    title_dict[title] += 1
```

```python
df = df[df['title'].notnull()][df["title"] != "Marked for Archive"]
```

```python
# df[df["title"] == "Marked for Archive"].head()
```

```python
df["title"].value_counts()
```

```python
df.dtypes
```

```python
title_asin_dict = defaultdict(int)
pks = []
for _, row in df.iterrows():
    try:
        cleaned_title = row["title"].lower().replace(" ", "-").replace(".", "").replace("\"", "").replace("'", "").replace("(", "").replace(")", "")
        pk = "-".join([cleaned_title, row["parent_asin"]])
        pks.append(pk)
        title_asin_dict[pk] += 1
    except:
        print(row)
    # title_asin_dict[
```

```python
len(title_asin_dict)
```

```python
len(pks)
```

Great, we can concatenate title and parent ASIN for a unique PK

```python
pks[:10]
```

```python
df["primary_key"] = pks
```

# Images

```python
import ast

def parse_images(image_str):
    try:
        parsed_list = ast.literal_eval(image_str)
        return parsed_list
    except Exception as e:
        print(f"Error parsing: {e}")
        return []
```

```python
# small_df["parsed_images"] = small_df["images"].apply(parse_images)
parsed_images = small_df["images"].apply(parse_images);
small_df.loc[:, "parsed_images"] = parsed_images
```

```python
import re
def extract_first_large_url(images_str: str) -> str:
    match = re.search(r"'large'\s*:\s*'([^']+)'", images_str)
    return match.group(1) if match else ""
```

```python
df["image_url"] = df["images"].apply(extract_first_large_url)
```

```python
image_urls_df = df[["primary_key", "image_url"]]
image_urls_df = image_urls_df.groupby("image_url")["primary_key"].apply(list).reset_index()
```
